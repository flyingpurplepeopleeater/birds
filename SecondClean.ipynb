Importing Modules

```python
import pandas as pd
import numpy as np
```

Doing additional work on the files from the first clean

```python
## Reading the text files from the first clean

table = pd.read_table("Data/amegfi/2019_ebd_amegfi_Feb-2024.txt", encoding = 'cp1252')
```

```python
print(table.info())

## Dropping all observations which did not report all the species (not reliable I guess?)

table = table[table['ALL SPECIES REPORTED'] == 1]

## Dropping redundant or useless columns

table1 = table.drop(['HAS MEDIA',
                      'APPROVED',
                      'REVIEWED',
                      'REASON',
                      'COUNTRY',
                      'COUNTY',
                      'STATE',
                      'ALL SPECIES REPORTED'], axis = 1)

print(table1.info())
```

```python
## Adjusting latitude and longitude by rounding down to the nearest whole number

table1['LATITUDE'] = (table1['LATITUDE']*10)//10
table1['LONGITUDE'] = (table1['LONGITUDE']*10)//10
table1.head(20)
```

```python
## Here I separated the observations that were made as a group (table_grp)
## and the observations that were made individually (table_ivd)

table2 = table1.copy()

grps = table2[table2['GROUP IDENTIFIER'].notnull()]
individs = table2.drop(grps.index, axis = 0)

print(grps.info())

print(grps['GROUP IDENTIFIER'].value_counts())



```

```python
## This is the first step at getting rid of duplicates (Group observations)

newidcol = {}
grpstoinds = {}
grpnames = []
badinds = []


for i in grps.index:
    groupid = grps['GROUP IDENTIFIER'][i]
    ismember = False
    date_error = False
    for name in grpnames:
        if name == groupid:
            ind = grpstoinds[groupid]
            newidcol[ind] += ('&' + grps.loc[i,'GLOBAL UNIQUE IDENTIFIER'])
            if grps.loc[ind, 'DAY'] != grps.loc[i, 'DAY']:
                date_error = True
            badinds.append(i)
            ismember = True
            break
    if ismember == False:
        grpstoinds[groupid] = i
        grpnames.insert(0,groupid)
        newidcol[i] = grps.loc[i,'GLOBAL UNIQUE IDENTIFIER']
        if len(grpnames) > 300:
            for name in grpnames[250:300]:
                del grpstoinds[name]
            del grpnames[250:300]

print(date_error)

```

```python
## The first step did some work

grps2 = grps.drop(badinds)
grps2['GLOBAL UNIQUE IDENTIFIER'] = newidcol
names2 = grps2['GROUP IDENTIFIER'].value_counts()
print(names2)
print (grps2['GLOBAL UNIQUE IDENTIFIER'].head())
```

```python
## This is the second step that completely removes all duplicates

bnames = []
for ind in names2.index:
    if names2[ind] > 1:
        bnames.append(ind)
print(len(bnames))

newidcol = {}
grpstoinds = {name : -1 for name in bnames}
badinds = []
date_error = False

for i in grps2.index:
    groupid = grps2['GROUP IDENTIFIER'][i]
    ismember = False
    for name in bnames:
        if name == groupid:
            if grpstoinds[name] == -1:
                grpstoinds[name] = i
                newidcol[i] = grps2.loc[i,'GLOBAL UNIQUE IDENTIFIER']
            else:
                ind = grpstoinds[groupid]
                newidcol[ind] += ('&' + grps2.loc[i,'GLOBAL UNIQUE IDENTIFIER'])
                if grps2.loc[ind,'DAY'] != grps2.loc[i,'DAY']:
                    date_error = True
                badinds.append(i)

print(date_error)

grps3 = grps2.drop(badinds)

for ind, uids in newidcol.items():
    grps3.loc[ind, 'GLOBAL UNIQUE IDENTIFIER'] = uids

names3 = grps3['GROUP IDENTIFIER'].value_counts()
bnames = []
for ind in names3.index:
    if names3[ind] > 1:
        bnames.append(ind)

print(bnames)
```

```python
print(grps3['GLOBAL UNIQUE IDENTIFIER'].head())
```

```python
## Some observations have 'X' in the observation count so we have to drop those

table3 = pd.concat([individs,grps3])

badinds = []

for i in table3.index:
    if table3.loc[i,'OBSERVATION COUNT'].isdigit() == False:
        badinds.append(i)
print(len(badinds))

```

```python
## Dropping columns that we are not going to use

table4 = table3.drop(badinds)
table4.info()

def is_in(sth, iterable):
    for item in iterable:
        if sth == item:
            return True
    return False

dropcols = [col for col in table4.columns if is_in(col,['OBSERVATION COUNT',
                                                        'OBSERVATION DATE',
                                                        'IBA CODE',
                                                        'BCR CODE',
                                                        'USFWS CODE',
                                                        'DURATION MINUTES',
                                                        'LATITUDE',
                                                        'LONGITUDE',
                                                        'YEAR',
                                                        'WEEK',
                                                        'DAY']) == False]

```

```python
## We create separate columns containing month, day, and year info
## We also simplify the information in the columns for IBA CODE, BCR CODE, and USFWS CODE

table5 = table4.drop(dropcols, axis = 1)
table5['IBA CODE'] = table5['IBA CODE'].notnull()
table5['BCR CODE'] = table5['BCR CODE'].notnull()
table5['USFWS CODE'] = table5['USFWS CODE'].notnull()

def get_mmdd(year,month, day):
    if (year < 1) or (day < 1):
        print('Error: wrong date')
        return None
    leap_year = False
    if year%4 == 0:
        leap_year = True
    if is_in(month, [4,6,9,11]):
        if day <= 30:
            return [month,day]
        else:
            return get_mmdd(year,month+1, day-30)
    elif month == 2:
        if day <= 28 + leap_year:
            return [month,day]
        else:
            return get_mmdd(year,month+1, day-28-leap_year)
    else:
        if day <= 31:
            return [month,day]
        else:
            return get_mmdd(year,month+1, day-31)

table5 = table5.rename(columns = {'WEEK' : 'MONTH'})

table5['MONTH'] = table5['DAY'].apply(lambda x: get_mmdd(2019,1,x)[0])
table5['DAY'] = table5['DAY'].apply(lambda x: get_mmdd(2019,1,x)[1])
table5['OBSERVATION COUNT'] = table5['OBSERVATION COUNT'].apply(lambda x: int(x))
table5 = table5.dropna()


print(table5.info())
print(table5.head(20))
```

```python
table5['COUNT'] = np.ones((len(table5['DAY']),1))

table5['INDEX'] = range(len(table5.index))

table6 = table5.set_index(['YEAR', 'MONTH', 'DAY', 'LATITUDE', 'LONGITUDE', 'INDEX']).sort_index()


print(table6.head(20))
```

```python
table6.to_csv("Data/amegfi/2019_ebd_amegfi_Feb-2024_secondclean.csv",
          index=True)
```
